## Q7. Sum of weights

Select all the cars from Asia

```python
asian_cars.head(2)
```

Select only columns vehicle_weight and model_year

```python
subset_asian_cars = asian_cars[['vehicle_weight', 'model_year']]; subset_asian_cars
```

Select the first 7 values

```python
first_7_records = subset_asian_cars[0:7]; first_7_records
```

Get the underlying NumPy array. Let's call it X.

```python
X = first_7_records.values; X
```

```python
print(f"XT shape: {X.shape}")
```

Compute matrix-matrix multiplication between the transpose of X and X.

```python
XT = X.T; XT
```

```python
print(f"XT shape: {XT.shape}")
```

```python
XTX = np.dot(XT, X)
```

```python
print(f"XTX shape: {XTX.shape}")
```

Invert XTX

```python
XTX_inverse = np.linalg.inv(XTX); XTX_inverse
```

```python
print(f"XTX_inverse shape: {XTX_inverse.shape}")
```

Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
print(f"y shape: {y.shape}")
```

Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.

```python
part1 = np.dot(XTX_inverse, XT) 
```

```python
w = np.dot(part1,y)
```

```python
w.shape
```

What's the sum of all the elements of the result?

```python
result_sum = round(np.sum(w),3); result_sum
```
