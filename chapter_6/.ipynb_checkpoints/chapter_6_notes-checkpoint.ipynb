## [6.1 Credit risk scoring project](https://www.youtube.com/watch?v=GJGmlfZoCoU&list=PL3MmuxUbc_hIhxl5Ji8t4O6lPAOpHaCLR&index=59)
Dataset: https://github.com/gastonstat/CreditScoring

```python
import pandas as pd
import numpy as np

from IPython.display import display

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import roc_auc_score
from sklearn.tree import export_text

from sklearn.ensemble import RandomForestClassifier
```

## [6.2 Data cleaning and preparation](https://www.youtube.com/watch?v=tfuQdI3YO2c&list=PL3MmuxUbc_hIhxl5Ji8t4O6lPAOpHaCLR&index=60)
- Downloading the dataset
- Re-encoding the categorical variables
- Doing the train/validation/test split

```python
csv_url = 'https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-06-trees/CreditScoring.csv'
```

```python
df = pd.read_csv(csv_url)
```

```python
df.head()
```

```python
## Data Cleaning Part
df.columns = df.columns.str.lower()
```

```python
# Mapping the variables back to the original values
status_values = {
    1: 'ok',
    2: 'default',
    0: 'unk'
}
df.status = df.status.map(status_values)
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk'
}

df.home = df.home.map(home_values)

marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'separated',
    5: 'divorced',
    0: 'unk'
}

df.marital = df.marital.map(marital_values)

records_values = {
    1: 'no',
    2: 'yes',
    0: 'unk'
}

df.records = df.records.map(records_values)

job_values = {
    1: 'fixed',
    2: 'partime',
    3: 'freelance',
    4: 'others',
    0: 'unk'
}

df.job = df.job.map(job_values)
```

```python
df.head()
```

```python
## Handling missing values
# Income, assets, and debt have large values, and the value 99999999 was used for replacing a missing value
df.describe().round()
```

```python
for c in ['income', 'assets', 'debt']:
    df[c] = df[c].replace(to_replace=99999999, value = np.nan)
```

```python
df.describe().round()
```

```python
# there is only 1 unknown record, so we should simply remove it
df.status.value_counts()
```

```python
df = df[df.status != 'unk'].reset_index(drop = True)
```

```python
## Train/test split
df_full_train, df_test = train_test_split(df, test_size = 0.2, random_state = 11)
df_train, df_val = train_test_split(df_full_train, test_size = 0.25, random_state = 11)
```

```python
df_train = df_train.reset_index(drop = True)
df_val = df_val.reset_index(drop = True)
df_test = df_test.reset_index(drop = True)
```

```python
y_train = (df_train.status == 'default').astype('int').values
y_val = (df_val.status == 'default').astype('int').values
y_test = (df_test.status == 'default').astype('int').values
```

```python
del df_train['status']
del df_val['status']
del df_test['status']
```

## [6.3 Decision trees](https://www.youtube.com/watch?v=YGiQvFbSIg8&list=PL3MmuxUbc_hIhxl5Ji8t4O6lPAOpHaCLR&index=61) = basically, a bunch of if-else statements
- How a decision tree looks like
- Training a decision tree
- Overfitting
- Controlling the size of a tree

```python
from PIL import Image
import matplotlib.pyplot as plt
img = Image.open('decision_tree.png')
plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# just a representation of the decision tree
def assess_risk(client):
    if client['records'] == 'yes':
        if client['job'] == 'parttime':
            return 'default'
        else:
            return 'ok'
    else:
        if client['assets']>6000:
            return 'ok'
        else:
            return 'default'
```

```python
xi = df_train.iloc[0].to_dict()
```

```python
assess_risk(xi)
```

**The rules of the data can be learned from the decision tree algo. So, let's use it!**

```python
# train data
train_dict = df_train.fillna(0).to_dict(orient='records')
dv = DictVectorizer(sparse = False)
X_train = dv.fit_transform(train_dict)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# test on validation dataset
val_dict = df_val.fillna(0).to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
# model predictions
y_pred = dt.predict_proba(X_val)[:, 1]
```

```python
# roc
roc_auc_score(y_val, y_pred)
```

```python
# auc for training dataset
y_pred = dt.predict_proba(X_train)[:, 1]
roc_auc_score(y_train, y_pred)
```

**Yay, overfitting = memoring the data but failing to generalize**

If we let our tree grow too deep, our model will fail to generalize. Instead of having over 7 levels of depth, we can use 3 levels of depth, the rules will be a bit more generalized.

```python
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict_proba(X_train)[:, 1]
auc = roc_auc_score(y_train, y_pred)
print('train auc: ', auc)

y_pred = dt.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print('val auc: ', auc)
```

**Restriction of the leaf growth had actually improved the model performance on the validation dataset by 8%**

*Terminology*: **decision stump** - a tree with one level of depth

```python
# Let's visualize the tree
# feature_names let's you access the actual names of the features
print(export_text(dt, feature_names=list(dv.get_feature_names_out())))
```

## [6.4 Decision tree learning algorithm](https://www.youtube.com/watch?v=XODz6LwKY7g&list=PL3MmuxUbc_hIhxl5Ji8t4O6lPAOpHaCLR&index=62)

```python
from PIL import Image
import matplotlib.pyplot as plt
img1 = Image.open('dt_terminology.png')
plt.figure(figsize=(10, 8))
plt.imshow(img1)
plt.axis('off')
plt.show()
```

```python
data = [
    [8000, 'default'],
    [2000, 'default'],
    [   0, 'default'],
    [5000, 'ok'],
    [5000, 'ok'],
    [4000, 'ok'],
    [9000, 'ok'],
    [3000, 'default'],
]

df_example = pd.DataFrame(data, columns=['assets', 'status'])
df_example
```

### Asset split

```python
from PIL import Image
import matplotlib.pyplot as plt
img1 = Image.open('split_decision1.png')
plt.figure(figsize=(10, 8))
plt.imshow(img1)
plt.axis('off')
plt.show()
```

```python
df_example.sort_values('assets')
```

```python
Ts = [0, 2000, 3000, 4000, 5000, 8000]
for T in Ts:
    print(T)
    df_left = df_example[df_example.assets <= T]
    df_right = df_example[df_example.assets > T]
    
    display(df_left)
    display(df_right)
    
    print()
```

### split evaluation criteria
Misclassification rate - predicting the same value of the split but being misclassfied

For T=4000, we get an average misclassification between assets > 4000 being true and false of 25%.
<br> **It is better to use a weighted average**

**impurity** is one of the ways to measure misclassificatin rate

```python
Ts = [0, 2000, 3000, 4000, 5000, 8000]
for T in Ts:
    print(T)
    df_left = df_example[df_example.assets <= T]
    df_right = df_example[df_example.assets > T]
    
    display(df_left)
    print(df_left.status.value_counts(normalize=True))
    display(df_right)
    print(df_right.status.value_counts(normalize=True))
    
    print()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
img2 = Image.open('best_split_example_data.png')
plt.figure(figsize=(10, 8))
plt.imshow(img2)
plt.axis('off')
plt.show()
```

T = 3000 has the weighted impurity of 10%

### How to do this with 2 features?

```python

data = [
    [8000, 3000, 'default'],
    [2000, 1000, 'default'],
    [   0, 1000, 'default'],
    [5000, 1000, 'ok'],
    [5000, 1000, 'ok'],
    [4000, 1000, 'ok'],
    [9000,  500, 'ok'],
    [3000, 2000, 'default'],
]

df_example = pd.DataFrame(data, columns=['assets', 'debt', 'status'])
df_example
```

```python
df_example.sort_values('debt')
```

```python
thresholds = {
    'assets': [0, 2000, 3000, 4000, 5000, 8000],
    'debt': [500, 1000, 2000]
}
```

```python
for feature, Ts in thresholds.items():
    print('#####################')
    print(feature)
    for T in Ts:
        print(T)
        df_left = df_example[df_example[feature] <= T]
        df_right = df_example[df_example[feature] > T]

        display(df_left)
        print(df_left.status.value_counts(normalize=True))
        display(df_right)
        print(df_right.status.value_counts(normalize=True))

        print()
    print('#####################')
```

```python
from PIL import Image
import matplotlib.pyplot as plt
img3 = Image.open('best_split_example_data_2_features.png')
plt.figure(figsize=(10, 8))
plt.imshow(img3)
plt.axis('off')
plt.show()
```

Finding the best split:
- find all thresholds for a feature
- split the dataset using F>T condition
- compute the impurity
- select the smallest value

### stopping criteria
- group is already pure
- tree reached depth limit
- group is too small to split

### decision tree algorithm
- fing the best split
- stop if max depth is reached
- if left is sufficiently large and not pure -- > repeat the algo
- if right is sufficiently large and not pure -- > repeat the algo

[More classification criteria](https://scikit-learn.org/stable/modules/tree.html#classification-criteria) 

## [6.5 Decision trees parameter tuning](https://www.youtube.com/watch?v=XJaxwH50Qok&list=PL3MmuxUbc_hIhxl5Ji8t4O6lPAOpHaCLR&index=64)
- Selecting `max_depth`
- Selecting `min_samples_leaf`

```python
depths = [1, 2, 3, 4, 5, 6, 10, 15, 20, None]

for depth in depths: 
    dt = DecisionTreeClassifier(max_depth=depth)
    dt.fit(X_train, y_train)
    
    y_pred = dt.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_pred)
    
    print('%4s -> %.3f' % (depth, auc))
```

The best values are 4,5,6. So, we'd use 4 in case if we are not tuning any other parameters

```python
scores = []

for d in [4, 5, 6]:
    for s in [1,2 ,5, 10, 15, 20, 100, 200, 500]:
        dt = DecisionTreeClassifier(max_depth=d, min_samples_leaf=s)
        dt.fit(X_train, y_train)

        y_pred = dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)

        scores.append((d, s, auc))
```

```python
columns = ['max_depth', 'min_samples_leaf', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
#df_scores.sort_values(by='auc', ascending = False)
```

```python
df_scores_pivot = df_scores.pivot(index='min_samples_leaf', columns=['max_depth'], values=['auc'])
```

```python
df_scores_pivot.round(3)
```

```python
# Visualize it as a heatmap - Lightest will give the best value
sns.heatmap(df_scores_pivot, annot=True, fmt=".3f")
```

```python
# We can experiment with more values just to see if we did not miss anything since our dataframe is small
scores = []

for d in [4, 5, 6, 7, 10, 15, 20, None]:
    for s in [1, 2 ,5, 10, 15, 20, 100, 200, 500]:
        dt = DecisionTreeClassifier(max_depth=d, min_samples_leaf=s)
        dt.fit(X_train, y_train)

        y_pred = dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)

        scores.append((d, s, auc))
```

```python
columns = ['max_depth', 'min_samples_leaf', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
df_scores.sort_values(by='auc', ascending = False).head()
```

```python
# Looks like a slightly better perfomance can be used with 10 max_depth and then min_samples_leaf of 15
```

```python
# Visualize it as a heatmap - Lightest will give the best value
sns.heatmap(df_scores_pivot, annot=True, fmt=".3f")
```

```python
## Train the final decision tree
dt = DecisionTreeClassifier(max_depth=6, min_samples_leaf=15)
dt.fit(X_train, y_train)
```

## 6.6 Ensembles and random forest
- Board of experts
- Ensembling models
- Random forest - ensembling decision trees
- Tuning random forest

```python
rf = RandomForestClassifier(n_estimators=10, random_state = 1)
rf.fit(X_train, y_train)
y_pred = rf.predict_proba(X_val)[:, 1]
```

```python
roc_auc_score(y_val, y_pred)
```

```python
rf.predict_proba(X_val[[0]])
```

```python
# Let's iterate over the number of estimators
scores = []

for n in range(10, 201, 10):
    rf = RandomForestClassifier(n_estimators=n, random_state=1)
    rf.fit(X_train, y_train)

    y_pred = rf.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_pred)
    
    scores.append((n, auc))
```

```python
df_scores = pd.DataFrame(scores, columns=['n_estimators', 'auc'])
```

## 6.10 Summary
- Decision trees learn if-then-else rules from data.
- Finding the best split: select the least impure split. This algorithm can overfit, that's why we control it by limiting the max depth and the size of the group.
- Random forest is a way of combininig multiple decision trees. It should have a diverse set of models to make good predictions.
- Gradient boosting trains model sequentially: each model tries to fix errors of the previous model. XGBoost is an implementation of gradient boosting.
