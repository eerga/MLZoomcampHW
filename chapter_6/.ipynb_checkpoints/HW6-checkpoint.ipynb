## Question 3

Now let's experiment with the n_estimators parameter

- Try different values of this parameter from 10 to 200 with step 10.
- Set random_state to 1.
- Evaluate the model on the validation dataset.
 
After which value of n_estimators does RMSE stop improving? Consider 3 decimal places for calculating the answer.

- 10
- 25
- 80
- 200

If it doesn't stop improving, use the latest iteration number in your answer.

```python
n_estimators_range = range(10, 201, 10)  # 10 to 200 with step 10
rmse_scores = []
best_rmse = float('inf')
best_n_estimators = 10
improvement_stopped_at = None

for n_est in n_estimators_range:
    # Train Random Forest
    rf = RandomForestRegressor(
        n_estimators=n_est,
        random_state=1,
        n_jobs=-1  # Add this for faster training
    )
    
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_val)
    
    # Calculate RMSE
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    rmse_scores.append((n_est, rmse))
    
    print(f"n_estimators={n_est:3d}: RMSE={rmse:.6f}")
    
    # Check if RMSE improved (considering 3 decimal places)
    if round(rmse, 3) < round(best_rmse, 3):
        best_rmse = rmse
        best_n_estimators = n_est
    else:
        # RMSE stopped improving (considering 3 decimal places)
        if improvement_stopped_at is None:
            improvement_stopped_at = best_n_estimators
            
print(f"\nBest RMSE: {best_rmse:.6f} at n_estimators={best_n_estimators}")

# Handle case where RMSE never stops improving
if improvement_stopped_at is None:
    improvement_stopped_at = best_n_estimators  # Use the latest iteration
    
print(f"RMSE stopped improving after n_estimators={improvement_stopped_at}")
```

```python
# Plot the results
import matplotlib.pyplot as plt
n_values, rmse_values = zip(*rmse_scores)
plt.figure(figsize=(10, 6))
plt.plot(n_values, rmse_values, marker='o')
plt.xlabel('n_estimators')
plt.ylabel('RMSE')
plt.title('RMSE vs n_estimators')
plt.grid(True)
plt.show()
```

After n_estimators=80 the RMSE stops improving significantly

## Question 4

Let's select the best `max_depth`:

- Try different values of `max_depth`: [10, 15, 20, 25]
- For each of these values,
    - try different values of `n_estimators` from 10 till 200 (with step 10)
    - calculate the mean RMSE
- Fix the random seed: `random_state=1`

What's the best max_depth, using the mean RMSE?

- 10
- 15
- 20
- 25

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

# Parameters to test
max_depth_values = [10, 15, 20, 25]
n_estimators_range = range(10, 201, 10)  # 10 to 200 with step 10

results = []

for max_depth in max_depth_values:
    rmse_scores = []
    
    print(f"\nTesting max_depth={max_depth}")
    
    for n_est in n_estimators_range:
        # Train Random Forest
        rf = RandomForestRegressor(
            n_estimators=n_est,
            max_depth=max_depth,
            random_state=1,
            n_jobs=-1
        )
        
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_val)
        
        # Calculate RMSE
        rmse = np.sqrt(mean_squared_error(y_val, y_pred))
        rmse_scores.append(rmse)
        
        print(f"  n_estimators={n_est:3d}: RMSE={rmse:.4f}")
    
    # Calculate mean RMSE for this max_depth
    mean_rmse = np.mean(rmse_scores)
    results.append({'max_depth': max_depth, 'mean_rmse': mean_rmse})
    print(f"  Mean RMSE for max_depth={max_depth}: {mean_rmse:.4f}")

# Convert to DataFrame for easy viewing
results_df = pd.DataFrame(results)
print("\n" + "="*50)
print("SUMMARY:")
print(results_df)

# Find the best max_depth
best_max_depth = results_df.loc[results_df['mean_rmse'].idxmin(), 'max_depth']
best_mean_rmse = results_df['mean_rmse'].min()

print(f"\nBest max_depth: {best_max_depth}")
print(f"Best mean RMSE: {best_mean_rmse:.4f}")
```

```python

# Visualize results
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(results_df['max_depth'], results_df['mean_rmse'])
plt.xlabel('max_depth')
plt.ylabel('Mean RMSE')
plt.title('Mean RMSE vs max_depth')
plt.grid(True, alpha=0.3)
for i, v in enumerate(results_df['mean_rmse']):
    plt.text(results_df['max_depth'].iloc[i], v + 0.001, f'{v:.4f}', ha='center')
plt.show()
```

max_depth 10 works the best

## Question 5

We can extract feature importance information from tree-based models.

At each step of the decision tree learning algorithm, it finds the best split. When doing it, we can calculate "gain" - the reduction in impurity before and after the split. This gain is quite useful in understanding what are the important features for tree-based models.

In Scikit-Learn, tree-based models contain this information in the [feature_importances_](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html#sklearn.ensemble.RandomForestRegressor.feature_importances_) field.

For this homework question, we'll find the most important feature:

- Train the model with these parameters:
    - `n_estimators=10`,
    - `max_depth=20`,
    - `random_state=1`,
    - `n_jobs=-1` (optional)
- Get the feature importance information from this model

What's the most important feature (among these 4)?

- `vehicle_weight`
- `horsepower`
- `acceleration`
- `engine_displacement`

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import pandas as pd

# Train Random Forest with specified parameters
rf = RandomForestRegressor(
    n_estimators=10,
    max_depth=20,
    random_state=1,
    n_jobs=-1
)

# Fit the model
rf.fit(X_train, y_train)

# Get feature importances
feature_importances = rf.feature_importances_

# Create a DataFrame to see feature names and their importances
# (assuming you have feature_names from your DictVectorizer or column names)
if hasattr(dv, 'feature_names_'):
    feature_names = dv.get_feature_names_out() 
else:
    # If you're using column names directly
    feature_names = X_train.columns if hasattr(X_train, 'columns') else [f'feature_{i}' for i in range(len(feature_importances))]

importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance': feature_importances
})

# Sort by importance (descending)
importance_df = importance_df.sort_values('importance', ascending=False)

print("Feature Importances:")
print(importance_df.head(10))

# Find the most important among the 4 specified features
target_features = ['vehicle_weight', 'horsepower', 'acceleration', 'engine_displacement']
target_importances = importance_df[importance_df['feature'].isin(target_features)]

print(f"\nImportances for target features:")
print(target_importances)

# Get the most important feature among the 4
most_important = target_importances.iloc[0]['feature']
print(f"\nMost important feature among the 4: {most_important}")
```

`vehicle_weight` is the most important one

## Question 6

Now let's train an XGBoost model! For this question, we'll tune the `eta` parameter:

- Install XGBoost
- Create DMatrix for train and validation
- Create a watchlist
- Train a model with these parameters for 100 rounds:

```python
#!pip install xgboost
```

```python
import xgboost as xgb
```

```python
features = list(dv.get_feature_names_out())
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
watchlist = [(dtrain, 'train'), (dval, 'val')]
```

```python
xgb_params = {
    'eta': 0.3, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}
```

```python
# Train model with watchlist
model = xgb.train(
    xgb_params, 
    dtrain, 
    num_boost_round=100,
    evals=watchlist,          
    verbose_eval=10        
)
```

```python
y_pred = model.predict(dval)
rmse = np.sqrt(mean_squared_error(y_val, y_pred))
print(f"Final RMSE: {rmse:.4f}")
```

Now change eta from 0.3 to 0.1.

Which eta leads to the best RMSE score on the validation dataset?

- 0.3
- 0.1
- Both give equal value


```python
xgb_params = {
    'eta': 0.1, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}
```

```python
# Train model with watchlist
model = xgb.train(
    xgb_params, 
    dtrain, 
    num_boost_round=100,
    evals=watchlist,          
    verbose_eval=10        
)
```

```python
y_pred = model.predict(dval)
rmse = np.sqrt(mean_squared_error(y_val, y_pred))
print(f"Final RMSE: {rmse:.4f}")
```

ETA of 0.1 gives a smaller RMSE, meaning it's a better parameter to use
